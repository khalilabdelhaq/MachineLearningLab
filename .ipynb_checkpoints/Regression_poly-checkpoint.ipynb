```python
import pandas as pd
```

```python
dataset = pd.read_csv("Regression_Lineaire_Simple/maison_localisation.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,2:3].values
```

```python
y = dataset.iloc[:,-1].values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y,marker= '+')
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
regressor = SGDRegressor()
```

```python
regressor.fit(X,y)
```

```python
plt.scatter(X,y,marker= '+')
plt.plot(X,regressor.predict(X),color='red')
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polymodel = PolynomialFeatures(degree=4)
```

```python
X_poly= polymodel.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
poly_regressor = LinearRegression()
```

```python
poly_regressor.fit(X_poly,y)
```

```python
plt.scatter(X,y,marker= '+',color='blue')
plt.plot(X,regressor.predict(X),color='red',label='Droite de la régression  linéaire')
plt.plot(X,poly_regressor.predict(X_poly),color='green',label = 'courbe de la régression polynomiale')
plt.legend()
```

```python
### Visualisation des modèles et apprentissage

Ce graphique permet de comparer le comportement de deux modèles après leur phase d’apprentissage :

- Les **points bleus (+)** représentent les données réelles du dataset  
- La **droite rouge** correspond à la régression linéaire  
- La **courbe verte** correspond à la régression polynomiale  

---

### Interprétation en termes d’apprentissage

- La **régression linéaire** apprend une relation simple (une droite).  
  → Elle peut être insuffisante si la relation réelle est non linéaire (**sous-apprentissage / underfitting**)

- La **régression polynomiale** apprend une relation plus complexe (courbe).  
  → Elle s’adapte mieux aux données si celles-ci sont non linéaires  

---

### Idée clé

L’apprentissage consiste à ajuster un modèle pour qu’il capture au mieux la relation entre les variables :

- Un modèle trop simple → ne capture pas bien les données  
- Un modèle trop complexe → risque de mémoriser le bruit (**surapprentissage / overfitting**)  

👉 Ce graphique permet donc de visualiser si le modèle a bien appris la structure des données.
```
