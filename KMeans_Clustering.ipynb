```python
import pandas as pd
```

```python
dataset = pd.read_csv('KMEANS/Mall_Customers.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for k in range(1,11):
    kmean= KMeans(n_clusters=k,init='k-means++',random_state=0)
    kmean.fit(X)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
import numpy as np
k_values = np.arange(1,11)
import matplotlib.pyplot as plt
plt.plot(k_values,wcss)
```

```python
kmean= KMeans(n_clusters=5,init='k-means++',random_state=0)
kmean.fit(X)
```

```python
y_clusters = kmean.predict(X)
```

```python
y_clusters
```

```python
plt.scatter(X[y_clusters == 0, 0],X[y_clusters == 0, 1],color="blue",label="Cluster1")
plt.scatter(X[y_clusters == 1, 0],X[y_clusters == 1, 1],color="orange",label="Cluster2")
plt.scatter(X[y_clusters == 2, 0],X[y_clusters == 2, 1],color="green",label="Cluster3")
plt.scatter(X[y_clusters == 3, 0],X[y_clusters == 3, 1],color="black",label="Cluster4")
plt.scatter(X[y_clusters == 4, 0],X[y_clusters == 4, 1],color="red",label="Cluster5")
plt.xlabel('Revenu annuel')
plt.ylabel('Score de dépense')
plt.legend()
```
